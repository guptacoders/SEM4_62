```python
from wordcloud import WordCloud,STOPWORDS
alice_novel = open("alice.txt",encoding="utf-8")
data = alice_novel.read()
print(data)
```

```python
stop = set(STOPWORDS)
stop
```

```python
# initialising object of wordcloud
alice_wl = WordCloud(background_color="white",max_words=2000,stopwords = stop)
alice_wl.generate(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,18))
plt.imshow(alice_wl)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016.csv")
df.head()
```

```python
df.shape
```

```python
df=df.iloc[0:100,:]
df
```

```python
import folium
latitude=37.77
longitude=-122.42
Sanfran_map=folium.Map(location=[latitude,longitude],zoom_start=12)
Sanfran_map
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng in zip(df.Y,df.X):
    incidents.add_child(folium.features.CircleMarker(location=[lat,lng],radius=5,color='Yellow',fill=True,fill_color="blue",fill_opacity=1))
Sanfran_map.add_child(incidents)
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng,label in zip(df.Y,df.X,df.Category):
    folium.Marker([lat,lng],popup=label).add_to(Sanfran_map)
Sanfran_map.add_child(incidents)
```
